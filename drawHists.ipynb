```python
import matplotlib.pyplot as plt
from coffea.util import load
from pocket_coffea.utils.plot_utils import Shape
from IPython.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
display(HTML("<style>.output_png { height: auto; }</style>"))
import os
if not os.path.exists('hists'):
    os.makedirs('hists')
```

```python
filename = "/Users/ashhome/Downloads/output_TTToSemiLeptonic_2018.coffea"
o = load(filename)
o.keys()
o['variables'].keys()
```

```python
hists = []
for var_name in o['variables'].keys():
    histname = var_name
    print(var_name)
    h_dict = o['variables'][histname]
    if histname == 'jet_eta_pt_leading': break;
    datamc = Shape(h_dict, histname, plot_dir='plots/')
    hists.append(datamc)
    #datamc
```

```python
for i,h in enumerate(hists):
    
    year = '2018'
    cat = '4b'
    h.build_stacks(year, cat)
    h.get_systematic_uncertainty()
    plt.figure() 
    h.stack_mc_nominal.plot(stack=False, histtype='fill')
    plt.legend()
    plt.savefig(f"hists/plot_{i}.png")
    #h.define_figure(year)
    #h.plot_systematic_uncertainty()
```
